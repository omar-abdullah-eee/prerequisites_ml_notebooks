#Set
1. Sets are unordered
2. Duplicate elements are not allowed
3. It starts with curly brackets

```python
fruits = {"mango","apple","watermelon","grapes","dragonfruit"};
print(fruits)

fruits = set(["mango","apple","watermelon","grapes","dragonfruit"]) #set() is a constructor
print(fruits)
print(type(fruits))
```

#accessing set elements

```python
fruits = {"mango","apple","watermelon","grapes","dragonfruit"};
print(fruits)
print("apple" in fruits)
print("banana" in fruits)
fruits.add("berry") #adding element in set
fruits.remove("mango") #removing elements from set
fruits.discard("uoo") #discarding elements from set
print(fruits)
```

##Dictionary


```python
student = {'name':'omar abdullah','age':26,'dept.':'EEE','Courses':["EEE121","EEE244"]};

print(student['name']) # accessing an element is dict

student["Courses"]= ["EEE 333"]; #replacing a element in dict
print(student)

#updating a dictionary
student.update(
    {'name': "abdullah"}
    )
print(student)

#deleting a dictionary
del student["dept."]
print(student)
```

#Positional arguments in function


```python
def names(first,second):
  print("First name:",first)
  print("Second name:",second)
  print("Full name",first,second)
  return

Student = names("Omar","Abdullah")
```

#Keyword argument(spcified order)

```python
def name(first,second):
  print("First name:",first)
  print("Second name:",second)
  print("Full name:",first,second)
  return

name(first="omar",second="Abdullah")
```
