##Arithmetic operation using np functions

```python
import numpy as np
a = np.array([1,2,4,2])
b = np.array([5,-2,4,2])
print(a)
print(b,"\n\n")

c = np.add(a,b)
print(c,"\n")

#substraction
c = np.subtract(a,b)
print(c,"\n")

#Multiplication
c = np.multiply(a,b)
print(c,"\n")

#Division
c = np.divide(a,b)
print(c,"\n")

#Modulo Division
c = np.mod(a,b)
print(c,"\n\n")

#power
c = np.power(a,2)
print(c,"\n")

#reciprocal
c = np.reciprocal(a)
print(c,"\n\n")

#Squire ROOT
c = np.sqrt(a)
print(c)
```

##Statistical arithmetic functions

```python
import numpy as np
x = np.array([1,2,1,3,4])
y = np.array([2,1,3,4,1])
print("Max value:",np.max(x))
print("Min value:",np.min(x))

#to find the bit position of the max and min value
print("Bit position for min number:",np.argmin(x))
print("Bit position for max number:",np.argmax(x),"\n\n")

#for multiple dimention array max,min calculation
#for axis = 0 : it will calc value according columns
#for axis = 1: it will calc value according rows

c = np.array([[4,3,7],[3,5,9]])
print("Max value:",np.max(c,axis=1))
print("Min value:",np.min(c,axis=0),"\n\n")

print("mean value:",np.mean(x))
print("Average value:",np.average(x))
print("Standard Deviation:",np.std(x))
print("Variance:",np.var(x))
print("Sum",np.sum(x))
print("Product:",np.prod(x))
print("Median:",np.median(x))
print("Cumulative Sum:",np.cumsum(x))
print("Cumulative Product:",np.cumprod(x),"\n\n")

#here the correlation coefficient is meansured as
#The correlation of x with x (which is always 1).
#The correlation of y with y (which is always 1).
#The correlation of x with y (-0.17647059).
#The correlation of y with x (-0.17647059), which is the same as x with y.
#equation
print("Correlation Coefficient:",np.corrcoef(x,y))
```

##Broadcasting numpy array

```python
import numpy as np
x = np.array([1,3])
y = np.array([[2,4],[4,4],[3,4],[5,9]])

#For NumPy broadcasting to work, arrays must have compatible shapes.
#Simply put, for two arrays to be broadcastable:
#Start from the trailing (rightmost) dimension: Compare the dimensions of the two arrays, moving from right to left.
#Compatibility Rule: For each dimension being compared, the dimensions must either:
#Be equal, or
#One of them must be 1.
#Missing Dimensions: If one array has fewer dimensions, its shape is effectively padded with leading ones (1) for comparison.

print(x.shape)
print(x)
print()
print("\n\n")
print(y.shape)
print(y)

print(x+y)
```

##Merging or joining arrays using concatenate

```python
import numpy as np
#1D arrays merging
#it creates two or multuple 1D arrays to one 1D array
x = np.array([1,2,3,4])
y = np.array([3,1,3,1])

mrg1 = np.concatenate((x,y))
print(x)
print(y)
print()
print(mrg1)

#2D arrays merging
a = np.array([[1,2],[3,5]])
b = np.array([[2,1],[3,9]])

mrg20 = np.concatenate((a,b),axis = 0) #merging according to column
print(mrg20)
print()

mrg21 = np.concatenate((a,b),axis = 1) #merging according to row
print(mrg21)
```

##Merging or joining arrays using stack

```python
import numpy as np
#stack creates two 1D arrays to 2D array by merging
#1D arrays merging
x = np.array([1,2,3,4])
y = np.array([3,1,3,1])

mrg10 = np.stack((x,y),axis = 0)
mrg11 = np.stack((x,y),axis = 1)
print("x:",x)
print("y:",y)
print()
print(mrg10) #which is a 2D array
print()
print(mrg11)
print()

#2D arrays merging
a = np.array([[1,2],[3,5]])
b = np.array([[2,1],[3,9]])

mrg20 = np.stack((a,b),axis = 0)
mrg20 = np.stack((a,b),axis = 0)

print("a:",a)
print("b:",b)
print()
mrg20 = np.concatenate((a,b),axis = 0) #merging according to column
print(mrg20)
print()

mrg21 = np.concatenate((a,b),axis = 1) #merging according to row
print(mrg21)
print()

#different approach to stack
mrg_horizontal = np.hstack((a,b))
mrg_vertical = np.vstack((a,b))
mrg_height = np.dstack((a,b)) #it can also be called dimention(d)

print(mrg_horizontal)
print()
print(mrg_vertical)
print()
print(mrg_height) #2D arrays to 3D array
```

##Flatten and Ravel

```python
import numpy as np
#both used as converting 2D to 1D array
#flatten data cant be altered as it works as np.copy
#Ravel data can be altered as it works as np.view

x = np.array(np.arange(9))
s = np.resize(x,(3,3))
print(s)
print()
#using flatten
l = s.flatten(order ="C")
s[0,1]= 99
print(l)
print()
#using Ravel
k = np.ravel(s)
print(s)
print()
print(k)
```

##Matrix functions in numpy

```python
import numpy as np
x = np.matrix([[1,2,3],[2,5,3],[9,7,3]])
print(x)
#Transpose
print("Transpose:\n",np.transpose(x))
#or
print()
print("Transpose:\n",x.T)

#swapaxes (same as transpose but swaps the axis of the matrix)
print()
print("Swapaxes:\n",np.swapaxes(x,0,1))
print()

#inverse matrix
print(np.linalg.inv(x))
print()

#power
print(np.linalg.matrix_power(x,5))
print()

#determinant
print(np.linalg.det(x))
print()

#diagonal fetching
print(np.linalg.diagonal(x))
```
